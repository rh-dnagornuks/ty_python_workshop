# Section 1 - Logical Operators

---
## Theory and Examples





Logical operators are used to combine conditional statements. Python has 3 of them:
- `and` - Returns True if both statements are true
- `or` - Returns True if one of the statements is true
- `not` - Reverses the result, returns False if the result is true

Example of `and` usage:

````python
a = 15
b = 20
if a > 10 and b < 25:
    print("Both conditions are True")
else:
    print("At least one of the conditions is False")
````

Example of `or` usage:

````python
a = 15
b = 20
if a > 10 or b > 25:
    print("At least one of the conditions is True")
else:
    print("None of the conditions are True")
````

Example of `not` usage:

````python
a = 15
if not a < 10:
    print("a is not less than 10")
````

You can combine multiple conditions using a mix of logical operators:

````python
hour = 19  # 7pm
has_badge = True
is_manager = False
is_security = False

after_hours = hour >= 18
allowed = has_badge and (not after_hours or is_manager or is_security)

print(f"Access to office allowed: {allowed}")
````

# Section 2 - Collections

---
## Theory and Examples




A **collection** is an object that groups many values together. Python’s built-in collection types differ in whether order matters, whether values can change, and how you access elements (by position or by name). Two main ones that will be covered are:

- **list** - ordered sequence written with square brackets `[...]`. Items can be added, removed, and changed. Duplicates are allowed.
- **set** - unordered collection of **unique** items. You write values in `{ ... }` without `key: value` pairs (unlike a dict). Membership tests (`in`) are natural with sets.

Example of a `list`:

````python
items = ["pen", "notebook", "pen"] # '[]' denote a list
items.append("eraser") # append(item) adds an item to the end of the list
print('The first item in the list is:', items[0]) # All lists begin with index 0. Use list[index] to access an item at a specific index.
print('The number of items in the list is:', len(items)) # len(collection) returns the number of items in the list
````

Example of a `set`:

````python
tags = {"python", "code", "python"}
print(tags) # Sets are unordered, so the order of items may change. Also, duplicates are discarded.
print("'code' tag is present:", "code" in tags) # Use 'in' to check if an item is in a collection
print("'python' tag is not present:", "python" not in tags) # Combine 'not' with 'in' to check if an item is not in a collection
````

**Note:** A literal `{ }` with no items is an **empty dictionary**, not an empty set. For an empty set, use `set()`.

# Section 3 - Loops

---
## Theory and Examples




**Loops** run the same block of code more than once. Python’s two main loop forms are:

- **`for`** - walks through items in a sequence (for example each element of a list) or another **iterable**. The loop variable takes on one value per repetition.
- **`while`** - repeats while a **condition** stays true. Use it when the number of repetitions is not known ahead of time, or when you want to stop based on a test you write yourself.

Helpful related tools:

- **`range(...)`** - built-in that yields a sequence of integers. It is often combined with `for` when you care about counting (0, 1, 2, …) or repeating an action a fixed number of times.

The body of a `for` or `while` loop must be **indented** under the line that starts with `for` or `while`.

Example of `for` over a `list`:

````python
tasks = ["email", "backup", "review"]
for task in tasks:
    print("Doing:", task)
````

Example of `for` with `range`:

````python
for i in range(1, 4): # range(a, b) returns a sequence of n numbers, starting from a and ending at b minus 1
    print("Count:", i)
````

Example of `while`:

````python
# Find largest multiple of 13 under 200
a = 199
while a % 13 != 0: # If a number is not a multiple of 13, it will have a remainder when divided by 13
    a = a - 1
print("Largest multiple of 13 under 200 is:", a)
````
