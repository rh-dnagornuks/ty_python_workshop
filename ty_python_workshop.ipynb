# Section 2 - Data Types and Variables

---
## Theory and Examples



Python has 4 basic data types:

- Integer [`int`]
- Floating point number (or decimal number) [`float`]
- String (or text) [`str`]
- Boolean (true or false) [`bool`]

Variables are containers of sorts that store a value (one of the data types). In Python, you use '=' to assign values to variable names.

```python
# Assigning values to variables
age = 22
height = 1.68
name = "Daniels"
student = True

# Referencing variables in a function
print("Name:", name)
print("Age:", age)

# Using function 'type()' to check the data type of a variable
print("Age:", type(age))
print("Height:", type(height))
print("Name:", type(name))
print("Student:", type(student))
```

You can also use arithmetic and assign new values to variables.

```python
num = 2
print("Old number: ", num)

num = num + 1
print("New number: ", num)
```

# Section 3 - Input

Use `input()` function to take input from the user and apply various operations to it.

```python
input("Enter any text to print it to the console: ")
```

You can save the input to a variable and include it in a print statement. You can also call input multiple times in a row.

```python
name = input("Enter your name: ")
place = input("Enter where you are from: ")
print(name, "is from", place)
```

# Section 4 - Comparison Operators

---
## Theory and Examples


Operators `==`, `!=`, `<`, `<=`, `>` and `>=` can be used to compare values and evaluate if the statement is `True` or `False`.

```python
print(1 < 2) # True
print(3 > 4) # False
```

You can mix comparison and arithmetic operators.

```python
print(1 + 11 >= 13)
print(2**4 < 30)
```

You can also use the operators to compare strings.
`==` and `!=` are used to check if words match.
`<`, `<=`, `>` and `>=` are used to check which word comes first alphabetically (similar to dictionary order).

```python
print('Apple' != 'Banana') # Different words, so True
print('Dog' < 'Cat') # 'Dog' comes after 'Cat', so False
```

# Section 5 - Conditionals

---
## Theory and Examples


Use `if` statement to run code if a condition is true.

```python
num = 5
if num > 4:
    print(num, 'is greater than 4')
```

You can use `else` statement to run code if the `if` statement doesn't run.

```python
num = 2
if num <= 3:
    print(num, 'is less than or equal to 3')
else:
    print(num, 'is greater than 3')
```

You can use `elif` (stands for 'else if') to chain multiple conditionals.

```python
num = 5
if num > 12:
    print(num, 'is greater than 12')
elif num < 12:
    print(num, 'is less than 12')
else:
    print(num, 'is equal to 12')
```

You can nest conditionals inside of other conditionals.


```python
num1 = 12
num2 = 0
if num2 != 0:
    if num1 % num2 == 0:
        print(num1, 'is evenly divisible by', num2)
    else:
        print(num1, 'is not evenly divisible by', num2)
else:
    print("Can't divide by zero")
```
